```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
# Changing the time to binary so that that the algorithm does not take a higher name as a more important feature
ohe_df = pd.DataFrame(ohe.fit_transform(df[['Time']]))
```

```python
ohe_df.isnull().sum()
```

```python
# resetting the index before/while joining will give us the dataframe we wont. If we didn't do this, each input will be turned into a NaN
new_df = pd.concat([df.reset_index(drop=True), ohe_df.reset_index(drop=True)], axis=1)
```
